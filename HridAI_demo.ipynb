```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)

time = np.arange(0, 100)
signal = np.random.rand(100)
spikes = signal > 0.8  # burst-like biochemical events

plt.figure(figsize=(10,3))
plt.plot(time, spikes.astype(int), drawstyle='steps-post')
plt.xlabel("Time")
plt.ylabel("Signal Event")
plt.title("Simulated Cellular Signaling Spikes")
plt.show()
```

```python
v = 0
threshold = 1.0
decisions = []

for s in spikes:
    v += s * 0.5          # signal accumulation
    v *= 0.9              # leakage
    decisions.append(1 if v > threshold else 0)
    if v > threshold:
        v = 0             # reset after activation
```

```python
plt.figure(figsize=(10,3))
plt.plot(decisions, drawstyle='steps-post')
plt.xlabel("Time")
plt.ylabel("Cell Activation")
plt.title("Neuro-Inspired Cellular Decision Output")
plt.show()
```

```python
reward = np.array(decisions) * 0.1  # beneficial activation
```

```python
weight = 0.5
weights = []

for r in reward:
    weight += 0.05 * r    # reward-modulated adaptation
    weights.append(weight)

plt.plot(weights)
plt.xlabel("Time")
plt.ylabel("Signal Sensitivity")
plt.title("Adaptive Plasticity (Learning Over Time)")
plt.show()
```
